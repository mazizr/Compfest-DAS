```python
import pandas as pd
```

```python
df = pd.read_csv("data-banjir-jakarta-3-tahun-terakhir.csv")
df = df.drop(df.columns[0], axis=1)
# df
```

```python
# Menggabungkan kolom kota_administrasi dan kelurahan dengan pemisah ", "
df['lokasi'] = df.apply(lambda row: row['kota_administrasi'] + ", " + row['kelurahan'], axis=1)
df = df.drop(df.columns[1], axis=1)
# Mengambil kolom 'lokasi' dan menyimpannya dalam variabel terpisah
lokasi = df.pop('lokasi')

# Menyisipkan kolom 'lokasi' ke posisi pertama dalam dfFrame
df.insert(2, 'lokasi', lokasi)
df['kota_administrasi'] = df['kota_administrasi'].str.lower()
df['kelurahan'] = df['kelurahan'].str.lower()
df['lokasi'] = df['lokasi'].str.lower()
df
```

```python
# Melakukan grouping berdasarkan kolom 'lokasi' dan menghitung jumlah dalam setiap grup
grouped_data = df.groupby('kota_administrasi').size()

# Mengurutkan hasil grouping dari yang terbesar ke terkecil
sorted_data = grouped_data.sort_values(ascending=False)

# Cetak hasil
print(sorted_data)
```

```python
# Filter hanya baris dengan string "Jakarta Utara" dalam kolom 'lokasi'
filtered_data = df[df['Lokasi'].str.contains('jakarta utara')]

# Cetak hasil
print(filtered_data)
```

```python
# Melakukan grouping berdasarkan kolom 'Lokasi' dan menghitung jumlah dalam setiap grup
grouped_data = filtered_data.groupby('Lokasi').size()

# Cetak hasil
print(grouped_data)
```
