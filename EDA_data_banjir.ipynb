```python
import pandas as pd
```

Data Banjir Jakarta 3 Tahun Terakhir

```python
df = pd.read_csv("data-banjir-jakarta-3-tahun-terakhir.csv")
df = df.drop(df.columns[0], axis=1)
# df
```

```python
# Menggabungkan kolom kota_administrasi dan kelurahan dengan pemisah ", "
df['lokasi'] = df.apply(lambda row: row['kota_administrasi'] + ", " + row['kelurahan'], axis=1)
df = df.drop(df.columns[1], axis=1)
# Mengambil kolom 'lokasi' dan menyimpannya dalam variabel terpisah
lokasi = df.pop('lokasi')

# Menyisipkan kolom 'lokasi' ke posisi pertama dalam dfFrame
df.insert(2, 'lokasi', lokasi)
df['kota_administrasi'] = df['kota_administrasi'].str.lower()
df['kelurahan'] = df['kelurahan'].str.lower()
df['lokasi'] = df['lokasi'].str.lower()
df['rw'] = df['rw'].str.lower()
df['rw'] = df['rw'].str.replace('rw', '')
df['rw'] = df['rw'].str.replace(':', '')
df.tail(10)
```

Grouping Wilayah

```python
tanggal_kejadian = df['tanggal_kejadian'].to_numpy()
# print(tanggal_kejadian)
for i in range(len(tanggal_kejadian)) :
    tahun = tanggal_kejadian[i].split()
    tanggal_kejadian[i] = tahun[-1]
    # print(tanggal_kejadian[i])
    
df['tahun_kejadian'] = tanggal_kejadian
df.tail(10)
```

```python
# Melakukan grouping berdasarkan kolom 'lokasi' dan menghitung jumlah dalam setiap grup
grouped_data = df.groupby('kota_administrasi').size()

# Mengurutkan hasil grouping dari yang terbesar ke terkecil
sorted_data = grouped_data.sort_values(ascending=False)

# Cetak hasil
print(sorted_data)
```

```python
# Filter hanya baris dengan string "Jakarta Utara" dalam kolom 'lokasi'
filtered_data = df[df['lokasi'].str.contains('jakarta utara')]

# Cetak hasil
print(filtered_data)
```

```python
# Melakukan grouping berdasarkan kolom 'Lokasi' dan menghitung jumlah dalam setiap grup
grouped_data = filtered_data.groupby('lokasi').size()

# Cetak hasil
print(grouped_data)
```

# Data Pasang Surut

```python
df_pasang_surut = pd.read_csv('data-pasang-surut-3-tahun-terakhir.csv')
df_pasang_surut['Tanggal'] = df_pasang_surut['Tanggal'].str.lower()
df_ps = df_pasang_surut.drop(df_pasang_surut.columns[[0,2,4]], axis=1)
df_ps.columns = df_ps.columns.str.lower()
df_ps
```

```python
# CEK TIPE DATA SETIAP KOLOM
df_ps.dtypes
```

```python
koef_angka = []
koef_status = [] 
koefisien = df_ps['koefisien'].to_numpy()
for i in range(len(koefisien)) :
    # String input
    string_input = str(koefisien[i])

    # Memisahkan angka dan kata menggunakan metode split()
    split_string = string_input.split()

    # Memisahkan angka dan kata
    angka = ""
    kata_biasa = []
    
    for elemen in split_string:
        if elemen.isdigit():
            angka = elemen
        else:
            kata_biasa.append(elemen)
            
    koef_angka.append(int(angka))
    koef_status.append(str(" ".join(kata_biasa)))
    # Output hasil pemisahan
    # print("Angka:", angka)
    # print("Kata Biasa:", " ".join(kata_biasa))

df_ps = df_ps.drop('koefisien', axis=1)
    
```

```python
from sklearn.preprocessing import LabelEncoder

df_ps['koefisien_angka'] = koef_angka
df_ps['koefisien_status'] = koef_status
# Buat kategorikal angka dari koef_status
status = df_ps['koefisien_status'].to_numpy()
koef_status_kategori = []
for i in range(len(status)) : 
    if status[i] == "s. tinggi" :
        koef_status_kategori.append(3)
    elif status[i] == "tinggi" :
        koef_status_kategori.append(2)
    elif status[i] == "rata-rata" :
        koef_status_kategori.append(1)
    elif status[i] == "rendah" :
        koef_status_kategori.append(0)
        
df_ps['koefisien_status_kategori'] = koef_status_kategori
df_ps
```

# Get Data Tanggal Banjir

```python
df.tail(50)
```

```python
import numpy as np 

tanggal_banjir = df['tanggal_kejadian'].to_numpy()
tanggal_banjir_jkt = []
for i in range(len(tanggal_banjir)) : 
    cek = tanggal_banjir[i].split()
    jumlah_cek = len(cek[:-2])
    if jumlah_cek > 1 : 
        for j in range(jumlah_cek) : 
            angka = cek[:-2]
            tanggal_banjir_jkt.append(str(angka[j][0:2]+" "+cek[-2:][0]+" "+cek[-2:][1]))
    else :
        tanggal_banjir_jkt.append(tanggal_banjir[i])
        
tanggal_banjir_jkt = [tanggal.lstrip('0') for tanggal in tanggal_banjir_jkt]
tanggal_banjir_jkt = [tanggal.lstrip() for tanggal in tanggal_banjir_jkt]
# tanggal_banjir_jkt[-10:]
            
```

```python
kondisi = []
kondisi_angka = []
tanggal_ps = df_ps['tanggal'].to_numpy()
for i in range(len(tanggal_ps)) : 
    for j in range(len(tanggal_banjir_jkt)) : 
        if tanggal_ps[i] == tanggal_banjir_jkt[j] : 
            ada_ga = True
            break
        else : 
            ada_ga = False
        
    if ada_ga == True :
        kondisi.append("banjir")
        kondisi_angka.append(1)
    else : 
        kondisi.append("tidak")
        kondisi_angka.append(0)
```

```python
df_ps["kondisi"] = kondisi
df_ps["kondisi_angka"] = kondisi_angka
df_ps.tail(50)
```
