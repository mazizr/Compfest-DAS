```python
import pandas as pd
import os
```

```python
from datetime import datetime

files = [file for file in os.listdir('./data-banjir-perbulan-2020')]

# Fungsi untuk mengubah string menjadi format tanggal
def sort_key(elem):
    # Mengambil kata pertama sebagai nama bulan
    bulan = elem.split()[0]
    
    # Mengubah nama bulan menjadi angka bulan dengan dictionary
    bulan_dict = {'januari': 1, 'februari': 2, 'maret': 3, 'april': 4, 'mei': 5, 'juni': 6, 'juli': 7, 'agustus': 8, 'september': 9, 'oktober': 10, 'november': 11, 'desember': 12}
    angka_bulan = bulan_dict[bulan]
    
    # Mengambil tahun dari string
    tahun = elem.split()[1].split('.')[0]
    
    # Mengembalikan tuple (tahun, bulan) untuk pengurutan
    return (tahun, angka_bulan)

# Mengurutkan list berdasarkan bulan
files = sorted(files, key=sort_key)

# # List comprehension untuk menambahkan string di setiap nilai
bulan = [item.split(".")[0] for item in files]
files = ["data-banjir-perbulan-2020/" + d for d in files]

bulan
```

```python
df = pd.DataFrame()
for i in range(len(files)) :
    df_data = pd.read_csv(files[i])
    df_data['ketinggian_air'] = df_data['ketinggian_air'].str.replace("s/d","-")
    df_data['tanggal_kejadian'] = df_data['tanggal_kejadian'].str.replace("tgl.", "").replace(" ", "")
    df_data['tanggal_kejadian'] = df_data['tanggal_kejadian']+" "+(str(bulan[i]))
    df = df.append(df_data)
    
df.to_csv("data-banjir-jakarta-2020.csv")
df
```
