```python
df_2020 = pd.read_csv("data-banjir-jakarta-2020.csv")
df_2020['tanggal_kejadian'] = df_2020['tanggal_kejadian'].str.lower()
# df_2020.tail(200)
```

```python
tanggal_kejadian_kotor = df_2020['tanggal_kejadian'].to_numpy()
tanggal_kejadian_bersih = []
for i in range(len(tanggal_kejadian_kotor)) :
    kata = tanggal_kejadian_kotor[i].split(" ")
    for j in range(len(kata)) : 
        if j > (len(kata)-1) : 
            continue
        if (kata[j]) == "" : 
            kata.pop(j)
    for k in range(len(kata)) :
        if k > (len(kata)-1) : 
            continue 
        if kata[k] == kata[k-1] : 
            kata.pop(k)
            k -= 1
        elif kata[k] == kata[k-2] :
            kata.pop(k)
            k -= 1
    cek = " ".join(kata).split(" ")
    if len(cek[0]) == 10 : 
        benerin = cek[0].split("-")
        benerin = benerin[2]
        cek[0] = benerin 
    tanggal_kejadian_bersih.append(" ".join(cek))

df_2020['tanggal_kejadian'] = tanggal_kejadian_bersih    
df_2020.to_csv("data-banjir-jakarta-2020.csv")
```
