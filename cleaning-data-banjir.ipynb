```python
import pandas as pd
```

```python
df = pd.read_csv("data-banjir-jakarta-3-tahun-terakhir.csv")
df = df.drop(df.columns[0], axis=1)
df = df.drop(df.columns[-1], axis=1)
# df = df.drop(['kecamatan','jumlah_terdampak_rw','jumlah_terdampak_rt','jumlah_terdampak_kk','lama_genangan','nilai_kerugian'], axis=1)
# df
```

```python
# Menggabungkan kolom kota_administrasi dan kelurahan dengan pemisah ", "
df['lokasi'] = df.apply(lambda row: row['kota_administrasi'] + ", " + row['kelurahan'], axis=1)
df = df.drop(df.columns[1], axis=1)
# Mengambil kolom 'lokasi' dan menyimpannya dalam variabel terpisah
lokasi = df.pop('lokasi')

# Menyisipkan kolom 'lokasi' ke posisi pertama dalam dfFrame
df.insert(2, 'lokasi', lokasi)
df['kota_administrasi'] = df['kota_administrasi'].str.lower()
df['kelurahan'] = df['kelurahan'].str.lower()
df['lokasi'] = df['lokasi'].str.lower()
df['rw'] = df['rw'].str.lower()
df['rw'] = df['rw'].str.replace('rw', '')
df['rw'] = df['rw'].str.replace(':', '')
df.head(10)
```

Clean Tangal

```python
tanggal_kejadian = df['tanggal_kejadian'].to_numpy()
tanggal_kejadian = [tanggal.lstrip() for tanggal in tanggal_kejadian]
for i in range(len(tanggal_kejadian)) : 
    tanggal_pisah = tanggal_kejadian[i].lower().split()
    if len(tanggal_pisah[0]) > 3 :
        sementara = tanggal_pisah[0].split(",")
        sementara = ", ".join(sementara)+" "+tanggal_pisah[-2:][0]+" "+tanggal_pisah[-2:][1]
        tanggal_pisah = sementara.split()

    # Menghapus koma dari setiap nilai dalam array
    tanggal_pisah = [element.replace(',', '') for element in tanggal_pisah]
    # Menghapus angka 0 di depan setiap nilai dalam array
    tanggal_pisah = [element.lstrip('0') for element in tanggal_pisah]
    tanggal_pisah = list(filter(bool, tanggal_pisah))
    tanggal_kejadian[i] = " ".join(tanggal_pisah)
    print(" ".join(tanggal_pisah))
    # for j in range(len(tanggal_pisah)) :
        
df['tanggal_kejadian'] = tanggal_kejadian
```

# Dipisah berdasarkan tanggal

```python
tanggal_kejadian = df['tanggal_kejadian'].to_numpy()
kota_administrasi = []	
kelurahan = []
lokasi = []
rw = []
jumlah_terdampak_jiwa = []
ketinggian_air = []
tanggal_banjir = []

for i in range(len(tanggal_kejadian)) :
    tanggal_asli = tanggal_kejadian[i].split()
    tanggal = tanggal_asli[:-2]
    if (len(tanggal)) == 1 : 
        kota_administrasi.append(df['kota_administrasi'].to_numpy()[i])	
        kelurahan.append(df['kelurahan'].to_numpy()[i])
        lokasi.append(df['lokasi'].to_numpy()[i])
        rw.append(df['rw'].to_numpy()[i])
        jumlah_terdampak_jiwa.append(df['jumlah_terdampak_jiwa'].to_numpy()[i])
        ketinggian_air.append(df['ketinggian_air'].to_numpy()[i])
        tanggal_banjir.append(df['tanggal_kejadian'].to_numpy()[i])
    else : 
        tanggal = [element.replace(',', '') for element in tanggal]
        for j in range(len(tanggal)) : 
            kota_administrasi.append(df['kota_administrasi'].to_numpy()[i])	
            kelurahan.append(df['kelurahan'].to_numpy()[i])
            lokasi.append(df['lokasi'].to_numpy()[i])
            rw.append(df['rw'].to_numpy()[i])
            jumlah_terdampak_jiwa.append(df['jumlah_terdampak_jiwa'].to_numpy()[i])
            ketinggian_air.append(df['ketinggian_air'].to_numpy()[i])
            tanggal_banjir.append(tanggal[j]+" "+tanggal_asli[-2:][0]+" "+tanggal_asli[-2:][1])
```

```python
# tanggal_banjir = [tanggal.lstrip() for tanggal in tanggal_banjir]
```

```python
df_banjir_bersih = pd.DataFrame({'tanggal_kejadian': tanggal_banjir, 
                   'kota_administrasi' : kota_administrasi,
                   'lokasi' : lokasi,
                   'rw' : rw, 
                   'jumlah_terdampak_jiwa' : jumlah_terdampak_jiwa, 
                   'ketinggian_air' : ketinggian_air})
df_banjir_bersih.tail(50)
df_banjir_bersih.to_csv("data-banjir-bersih.csv")
```
